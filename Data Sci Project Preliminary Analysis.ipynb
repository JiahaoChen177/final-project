```python
import pandas as pd
import numpy as np
import os
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('NTN-data.csv')
df
```

```python
df.describe().T
```

```python
fig, ax = plt.subplots()

ax.scatter(df['pH'], df['SO4'], label = 'SO4');
ax.scatter(df['pH'], df['NO3'], label = 'NO3');
ax.scatter(df['pH'], df['Ca'], label = 'Ca');
ax.scatter(df['pH'], df['Mg'], label = 'Mg');
ax.scatter(df['pH'], df['K'], label = 'K');
ax.scatter(df['pH'], df['Na'], label = 'Na');
ax.scatter(df['pH'], df['Cl'], label = 'Cl');
ax.scatter(df['pH'], df['NH4'], label = 'NH4');

ax.set_title('Ion Concentration vs pH');
ax.set_xlabel('pH');
ax.legend();
ax.set_ylabel('Ion Concentrations (mg/L)');
plt.ylim(0,5)
```

```python
fig, ax = plt.subplots()

ax.scatter(df['cum month'], df['SO4'], label = 'SO4');
ax.scatter(df['cum month'], df['NO3'], label = 'NO3');

ax.set_title('Ion Concentration Over Time');
ax.set_xlabel('Cumulative Months from 1999-2020');
ax.legend();
ax.set_ylabel('Ion Concentrations (mg/L)');
plt.ylim(0,5)
```

```python
fig, ax = plt.subplots()

ax.scatter(df['cum month'], df['pH']);

ax.set_title('pH Over Time');
ax.set_xlabel('Cumulative Months from 1999-2020');
ax.set_ylabel('pH');
```
